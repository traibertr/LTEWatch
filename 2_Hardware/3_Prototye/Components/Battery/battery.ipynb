```python
ic_part_number      = ["RJD2032C1","RJD2048","RJD2430C1","RJD2440","RJD2450","RJD3032","RJD3048","RJD3555"]
capa                = [80,110,104,140,190,190,290,512]
i_charging          = [40,55,53,70,95,95,145,275.5]
i_discharging_max   = [160,220,208,380,380,580,1030]
weight              = [3.4,4.2,4.5,5.4,6.5,7.2,9.3,14.1]
max_diameter        = [20.02,20.02,24.5,24.5,24.5,30,30,35.2]
thickness           = [3.5,5.0,3.15,4.3,5.4,3.4,4.9,5.7]

price               = [6.6779,7.6636,7.5863,8.8909,7.9342,7.7602,8.0502,19.5214] # Price from futureelectronics.com

capa                = np.array(capa)
i_charging          = np.array(i_charging)
i_discharging_max   = np.array(i_discharging_max)
weight              = np.array(weight)
max_diameter        = np.array(max_diameter)
thickness           = np.array(thickness)
price               = np.array(price)

volume              = math.pi*(np.array(max_diameter)/2)**2
```

```python
# Price and Capacity
plt.plot(capa,price,'--+',linewidth=1)
plt.legend
plt.ylabel("Price [CHF]")
plt.xlabel("Min Capacity [mAh]")
plt.grid()
plt.title("Price and Capacity")
plt.xlim([50,550])
plt.ylim([0,22])

# Power densitiy (Wh/kg, Wh/l, Wh/CHF)

volume = volume/10**(-6)
power = capa*3.7/1000           # Wh = mAh * V / 1000

power_density_1 = power/weight
power_density_2 = power/volume
power_density_3 = power/price

fig, ax = plt.subplots()

plt.stem(capa,power_density_1)
plt.stem(capa,power_density_2)
plt.stem(capa,power_density_3)
plt.legend
plt.ylabel("Power Density")
plt.xlabel("Capacity [mAh]")
plt.grid()
plt.title("Power densities")
plt.xlim([50,550])


```
